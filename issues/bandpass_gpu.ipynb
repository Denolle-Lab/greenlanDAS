```julia
using SeisNoise, Plots, CUDA
import SeisIO.get_data
```

```julia
# download data
S = get_data("FDSN","CI.SDD..BHZ,CI.PER..BHZ",src="SCEDC",s="2019-02-03T00",t="2019-02-03T01")
S1 = deepcopy(S)
S2 = deepcopy(S)
```

```julia
# set sampling frequency and filter band in Hz
freqmin,freqmax = 0.1,1

# set correlation step, length and max lag
cc_step, cc_len = 450, 1800
maxlag = 60.
```

```julia
# preprocess data on cpu
process_raw!(S1,S1.fs[1])
Raw1 = RawData.([S1[1],S1[2]],cc_len,cc_step)
detrend!.(Raw1)
taper!.(Raw1)
bandpass!.(Raw1,freqmin,freqmax,zerophase=true)

# correlate on cpu
FFT1 = rfft.(Raw1)
whiten!.(FFT1,freqmin,freqmax)
C1 = correlate(FFT1[1],FFT1[2],maxlag)

# process correlation function
bandpass!(C1,freqmin,freqmax)
abs_max!(C1)
```

```julia
# preprocess on gpu
process_raw!(S2,S2.fs[1])
R1 = RawData(S2[1], cc_len, cc_step) |> gpu
R2 = RawData(S2[2], cc_len, cc_step) |> gpu
Raw2 = [R1,R2]
detrend!.(Raw2)
taper!.(Raw2)
bandpass!.(Raw2,freqmin,freqmax,zerophase=true)

# correlate on gpu
FFT2 = rfft.(Raw2)
whiten!.(FFT2,freqmin,freqmax)
C2 = correlate(FFT2[1],FFT2[2],maxlag) |> cpu

# process correlation function
bandpass!(C2,freqmin,freqmax)
abs_max!(C2)
```

```julia
plot([C1.corr[:,1],C2.corr[:,1]])
```
