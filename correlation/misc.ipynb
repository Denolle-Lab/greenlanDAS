```julia
# plot each leg of the correlation functions

# get just the first leg of correlations to check cross-cable stacking
indices = [j for j in combinations(collect(chans[1]:chans[2]),2)]
indices = reduce(vcat,transpose.(indices))
leg1_indices = (indices[:,1] .<= midpoint .&& indices[:,2] .<= midpoint)
C_leg1 = deepcopy(C_filt)
C_leg1.corr = C_filt.corr[:,leg1_indices]

# plot results
midpoint = Int64(chans[1]+(chans[2]+1-chans[1])/2-1)
plot_chans = [chans[1],midpoint]
title = string("Leg 1 correlation functions at ",freqmin,"-",freqmax,
           "Hz (virtual source at channel ",source_chan,")")
fname = string("figures/correlations/fk_",cmin,"_",cmax,"/fk_",sgn,"/1khz_stack_leg1_",freqmin,"-",freqmax,
    "Hz_source_",source_chan,".png")
plot_correlations(C_leg1,"common shot",maxlag,plot_chans,source_chan,330,title,fname)

# second leg
i = sortperm(indices[:,2])
sorted_indices = reverse(indices[i,:])
sorted_corr = reverse(C_filt.corr[:,i])
leg2_indices = (sorted_indices[:,1] .> midpoint .&& sorted_indices[:,2] .> midpoint)
C_leg2 = deepcopy(C_filt)
C_leg2.corr = sorted_corr[:,leg2_indices]

# plot results
midpoint = Int64(chans[1]+(chans[2]+1-chans[1])/2-1)
plot_chans = [chans[1],midpoint]
title = string("Leg 2 correlation functions at ",freqmin,"-",freqmax," Hz (virtual source at surface)")
fname = string("figures/correlations/fk_",cmin,"_",cmax,"/fk_",sgn,"/1khz_stack_leg2_",freqmin,"-",freqmax,
               "Hz_source_",source_chan,".png")
plot_correlations(C_leg2,"common shot",maxlag,plot_chans,source_chan,330,title,fname)

# third leg
i = sortperm(indices[:,2])
sorted_indices2 = reverse(indices[i,:])
sorted_corr2 = reverse(C_filt.corr[:,i])
i = sortperm(sorted_indices2[:,2])
sorted_indices2 = reverse(sorted_indices2[i,:],dims=2)
sorted_corr2 = reverse(sorted_corr2[:,i],dims=1)
cross_indices = (sorted_indices2[:,1] .<= midpoint .&& sorted_indices2[:,2] .> midpoint)
leg3_indices = (cross_indices + (reduce(vcat,sum(sorted_indices2,dims=2)) .<= chans[end]+chans[1]-1) .== 2)
C_leg3 = deepcopy(C_filt)
C_leg3.corr = sorted_corr2[:,leg3_indices]

# plot results
midpoint = Int64(chans[1]+(chans[2]+1-chans[1])/2-1)
plot_chans = [chans[1],midpoint]
title = string("Leg 3 correlation functions at ",freqmin,"-",freqmax," Hz (virtual source at surface)")
fname = string("figures/correlations/fk_",cmin,"_",cmax,"/fk_",sgn,"/1khz_stack_leg3_",freqmin,"-",freqmax,
               "Hz_source_",source_chan,".png")
plot_correlations(C_leg3,"common shot",maxlag,plot_chans,source_chan,330,title,fname)

# fourth leg
i = sortperm(indices[:,1])
sorted_indices3 = reverse(indices[i,:])
sorted_corr3 = reverse(C_filt.corr[:,i])
i = sortperm(sorted_indices3[:,1])
sorted_indices3 = reverse(sorted_indices3[i,:])
sorted_corr3 = reverse(sorted_corr3[:,i])
cross_indices = (sorted_indices3[:,1] .<= midpoint .&& sorted_indices3[:,2] .> midpoint)
leg4_indices = (cross_indices + (reduce(vcat,sum(sorted_indices3,dims=2)) .> chans[end]+chans[1]) .== 2)
C_leg4 = deepcopy(C_filt)
C_leg4.corr = reverse(sorted_corr3[:,leg4_indices],dims=1)

# plot results
midpoint = Int64(chans[1]+(chans[2]+1-chans[1])/2-1)
plot_chans = [chans[1],midpoint]
title = string("Leg 4 correlation functions at ",freqmin,"-",freqmax," Hz (virtual source at surface)")
fname = string("figures/correlations/fk_",cmin,"_",cmax,"/fk_",sgn,"/1khz_stack_leg4_",freqmin,"-",freqmax,
               "Hz_source_",source_chan,".png")
plot_correlations(C_leg4,"common shot",maxlag,plot_chans,source_chan,330,title,fname)
```
