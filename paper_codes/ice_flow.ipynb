```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# slope of bedrock obstacle
alpha = 1

# horizontal velocity
v = 1

# ice thickness
H = 1100

# grid
x = np.linspace(0,1,100)
z = np.linspace(0,H,100)
X,Z = np.meshgrid(x,z)

# set bedrock bump size
k = 1/100

# calculate stresses and strain rate
factor = 2.0*alpha*k**2*v*H * (Z/H-1)
tzz =  factor * np.cosh(-k*Z)
txz = factor * np.sinh(-k*Z)
tE = np.sqrt((2*tzz**2 + txz**2)/2)

# plot stress 
c=plt.contour(x/k,z,tzz*np.sin(2*np.pi*X)/(np.max(np.abs(tzz))),
                     cmap='seismic',vmin=-1,vmax=1)

# time dependence
np.exp(1j t)

plt.xlabel('Distance along flow(m)')
plt.ylabel('Depth (m)')
plt.title(f'$\lambda$={round(1/k)} m')
plt.gca().invert_yaxis()
plt.show()

```
