```julia
using SeisIO, SeisNoise, Plots, CUDA
```

```julia
# download data
S = get_data("FDSN","CI.SDD..BHZ,CI.PER..BHZ",src="SCEDC",s="2019-02-03T00",t="2019-02-04T01")
S1 = deepcopy(S)
S2 = deepcopy(S)
```

```julia
# set sampling frequency and filter band in Hz
fs = 40. 
freqmin,freqmax = 0.1,0.3

# set correlation step, length and max lag
cc_step, cc_len = 450, 1800
maxlag = 60.
smoothing_half_win = 12
```

```julia
# preprocess data on cpu
process_raw!(S1,fs)
Raw1 = RawData.([S1[1],S1[2]],cc_len,cc_step)
detrend!.(Raw1)
taper!.(Raw1)
bandpass!.(Raw1,freqmin,freqmax,zerophase=true)

# correlate on cpu
FFT1 = rfft.(Raw1)
whiten!.(FFT1,freqmin,freqmax)
C1 = correlate(FFT1[1],FFT1[2],maxlag)

# process correlation function
bandpass!(C1,freqmin,freqmax)
abs_max!(C1)

# plot correlation function
plot(C1)
```

```julia
# preprocess on gpu
R1 = RawData(S2[1], cc_len, cc_step) |> gpu
R2 = RawData(S2[2], cc_len, cc_step) |> gpu
Raw2 = [R1,R2]
detrend!.(Raw2)
taper!.(Raw2)
bandpass!.(Raw2,freqmin,freqmax,zerophase=true)

# correlate on gpu
FFT2 = rfft.(Raw2)
whiten!.(FFT2,freqmin,freqmax)
C2 = correlate(FFT2[1],FFT2[2],maxlag) |> cpu

# process correlation function
bandpass!(C2,freqmin,freqmax)
abs_max!(C2)

# plot correlation function
plot(C2)
```

```julia
# download data
S = get_data("FDSN","CI.SDD..BHZ,CI.PER..BHZ",src="SCEDC",s="2019-02-03T00",t="2019-02-03T01")
S1 = deepcopy(S)
S2 = deepcopy(S)

# set frequency band
freqmin,freqmax = 0.1,0.3
cc_step, cc_len = 450, 1800
```

```julia
# filter on cpu
R1 = RawData(S1[1],cc_len,cc_step)
bandpass!(R1,freqmin,freqmax,zerophase=true)
print(sum(R1.x),"\n")
plot(R1.x[:,1])
```

```julia
# filter on gpu
R2 = RawData(S2[1],cc_len,cc_step)
R2 = R2 |> gpu
bandpass!(R2,freqmin,freqmax,zerophase=true)
R2 = R2 |> cpu
print(sum(R2.x),"\n")
plot(R2.x[:,1])
```

```julia
# detrend and taper gives different results on GPU vs CPU?
function get_S(sz,fs)
    Random.seed!(1234)
    data = rand(sz,sz)
    S = SeisNoise.NodalData()
    S.data = data
    S.fs = ones(sz).*fs
    return S
end

function get_C(sz,fs)
    Random.seed!(1234)
    data = rand(sz,sz)
    C = SeisNoise.CorrData(corr=data,fs=fs)
    return C
end

# choose parameters
sz,fs = 100,10.
freqmin,freqmax = 1,4

# detrend on cpu
S1 = get_S(sz,fs)
detrend!(S1)
print(sum(S1.data),"\n")

# detrend on gpu
S2 = get_S(sz,fs) |> gpu
detrend!(S2)
S2=S2|>cpu
print(sum(S2.data),"\n")

# taper on cpu
S3 = get_S(sz,fs)
taper!(S3)
print(sum(S3.data),"\n")

# taper on gpu
S4 = get_S(sz,fs) |> gpu
taper!(S4)
S4 = S4|>cpu
print(sum(S4.data),"\n")

# bandpass on cpu
S5 = get_C(sz,fs)
SeisNoise.bandpass!(S5,freqmin,freqmax,zerophase=true)
print(sum(S5.corr),"\n")

# bandpass on gpu
S6 = get_C(sz,fs) |> gpu
SeisNoise.bandpass!(S6,freqmin,freqmax,zerophase=true)
S6 = S6|>cpu
print(sum(S6.corr),"\n")
```
